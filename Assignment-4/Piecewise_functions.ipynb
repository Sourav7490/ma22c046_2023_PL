```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.linspace(-5,5,1000001)
```

```python
def f(x):
     return np.where(x<0,x*np.sin(x**2),np.cos(x))
print(x)
```

```python
function_values=f(x)
```

```python
plt.figure(figsize=(10,6))
plt.xlabel('x')
plt.plot(x,function_values)
plt.ylabel('f(x)')
plt.title("Plot of the Piecewise Function")
plt.grid(True)
```
