```python
import urllib
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import shutil
```

```python
url="https://gml.noaa.gov/webdata/ccgg/trends/co2/co2_mm_mlo.txt"
co2folder='../../data_external/co2'
#downloading file
```

```python
if not os.path.isdir('../../data_external/'):
  os.mkdir('../../data_external/')
r = urllib.request.urlopen(url)
fo = open(co2folder, 'wb')
shutil.copyfileobj(r, fo)
fo.close()
```

```python
lines = open(co2folder, 'r').readlines()
```

```python
time=[] ; avg=[]
```

```python
for line in lines:
    if line.startswith('#') or len(line.strip())==0:
        continue  # skipping comments and empty lines

    data=line.split()

    time.append(float(data[2]))
    avg.append(float(data[3]))
```

```python
plt.figure(figsize=(8,6))
plt.plot(time,avg)
plt.xlabel('Time')

plt.ylabel('Monthly Avarage of CO2')
plt.title('Monthly Average CO2 Levels')
plt.show()
```

```python
rate_of_change= np.gradient(avg,time)
```

```python
plt.figure(figsize=(8,5))
plt.xlabel('Time')
plt.ylabel('Rate of change of C02')
plt.plot(time,rate_of_change, marker='')
plt.show()
```
